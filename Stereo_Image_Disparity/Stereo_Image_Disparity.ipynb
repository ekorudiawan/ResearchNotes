```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images\images1.jpg')
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
height, width = rgb.shape[:2]
print height
print width
```

```python
left_img = rgb[0:386, 0:499]
right_img = rgb[0:386, 501:1000]
```

```python
plt.imshow(left_img)
plt.title('Left Image')
plt.show()
```

```python
plt.imshow(right_img)
plt.title('Right Image')
plt.show()
```

```python
height, width = left_img.shape[:2]
print height
print width
```

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=33)
disparity = stereo.compute(left_img,right_img)
plt.imshow(disparity,'gray')
plt.show()
```
