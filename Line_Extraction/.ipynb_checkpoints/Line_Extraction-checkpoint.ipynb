```python
import time
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
import ipywidgets as wg
from IPython.display import display

```

```python
# Definisi umum
IMAGE_WIDTH = 640
HALF_IMAGE_WIDTH = IMAGE_WIDTH / 2
IMAGE_HEIGHT = 360
```

```python
# ini masukkan ke fungsi
angleStep = 1
lengthStep = 5

iepfEnable = True
oneCylce = False
showBinary = False
```

```python
# Transform scanning coordinat to camera coordinat
def transToImgFrame(x,y):
    # x = x + 160
    x = x + HALF_IMAGE_WIDTH
    y = (y-IMAGE_HEIGHT) * -1
    return (x,y)
```

```python
# Polar to cartesian 
def polToCart(radius, theta):
    x = int(radius * math.cos(math.radians(theta)))
    y = int(radius * math.sin(math.radians(theta)))
    return (x,y)
```

## Load Image From File

```python
matplotlib.rcParams['figure.figsize'] = (15,15)
```

```python
rgb_image = cv2.imread('images\image4.jpg')
rgb_image = cv2.resize(rgb_image,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)

plt.imshow(rgb_image)
plt.title('RGB Image')
plt.show()
```

## Convert RGB Image to HSV Image

```python
hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2HSV)

plt.imshow(hsv_image)
plt.title('HSV Image')
plt.show()
```

## Create Binary Image

```python
# Setting HSV Green
gHMin = wg.IntSlider(min=0, max= 255, description='Gr H Min : ')
gSMin = wg.IntSlider(min=0, max= 255, description='Gr S Min : ')
gVMin = wg.IntSlider(min=0, max= 255, description='Gr V Min : ')

gHMax = wg.IntSlider(min=0, max= 255, description='Gr H Max : ')
gSMax = wg.IntSlider(min=0, max= 255, description='Gr S Max : ')
gVMax = wg.IntSlider(min=0, max= 255, description='Gr V Max : ')

display(gHMin, gSMin, gVMin, gHMax, gSMax, gVMax)
```

```python
lowerGreen = np.array([gHMin.value,gSMin.value,gVMin.value])
upperGreen = np.array([gHMax.value,gSMax.value,gVMax.value])
gBinary = cv2.inRange(hsv_image, lowerGreen, upperGreen)
plt.imshow(gBinary, cmap='Greys')
plt.title('Green Binary Image')
plt.show()
```

```python
# Setting HSV White
wHMin = wg.IntSlider(min=0, max= 255, description='Wh H Min : ')
wSMin = wg.IntSlider(min=0, max= 255, description='Wh S Min : ')
wVMin = wg.IntSlider(min=0, max= 255, description='Wh V Min : ')

wHMax = wg.IntSlider(min=0, max= 255, description='Wh H Max : ')
wSMax = wg.IntSlider(min=0, max= 255, description='Wh S Max : ')
wVMax = wg.IntSlider(min=0, max= 255, description='Wh V Max : ')

display(wHMin, wSMin, wVMin, wHMax, wSMax, wVMax)
```

```python
lowerWhite = np.array([wHMin.value,wSMin.value,wVMin.value])
upperWhite = np.array([wHMax.value,wSMax.value,wVMax.value])
wBinary = cv2.inRange(hsv_image, lowerWhite, upperWhite)
plt.imshow(wBinary, cmap=plt.cm.binary)
plt.title('White Binary Image')
plt.show()
```
